# Bring out the OpenStack data

```python
# Load the data, this is an expensive operation.
df = pd.read_csv(
    'openstack-data.csv', 
    low_memory=False, 
    parse_dates=[
        'created', 'created.local', 'merged.date', 'merged.date.local', 
        'Workflow.approved.on', 'Workflow.approved.on.local'],
    date_parser=lambda x: pd.to_datetime(x, coerce=True),
    index_col='created',
    encoding='utf-8',
)
```

- This file is generated by a Python script I have that iterates through the gerrit API and actually uses pandas to create the CSV.
- A bunch of options
- low_memory=False is more flexible
- I've told it where the dates are
- utf-8, mostly for names around the world
- Index by the date as that interests me most at the moment

# How many reviews does a typical patch get?

```python
incubated_projects = df[df['project.group'] == 'openstack']
reviewed = incubated_projects.groupby(['Code-Review.total'])
reviewed['_number'].count().plot(xticks=range(0, 70, 2))
```

- Start with getting a feel for the data and checking things look right
- I have said to people how many reviews you typically get
- higher than projects without the 2x+2 requirement

# Theory: Most people work Mondy - Friday

```python
# Which day is the most popular for submitting reviews? (localised)
df_local = df.set_index('created.local')
df_local['weekday'] = df_local.index.weekday
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Friday']
counts_by_day = df_local[df_local.weekday >= 0 ].groupby('weekday').count()
counts_by_day.index = days
counts_by_day['_number'].plot(kind='bar')
```

- Most people working on OpenStack do so for work
- Probably different on other Open Source projects
- However, still a fair bit of activity at the weekend!

Steps:
- reindex
- create a weekday col from another
- group by new col
- count, add labels and plot

# Theory: Most people work 9-5 on OpenStack

```python
df['hour'] = df.index.hour
counts_by_hour = df[df.hour >= 0 ].groupby('hour')
counts_by_hour['_number'].count().plot(kind='bar')
```

# Theory: Most people work 9-5 on OpenStack (localised)

```python
df_local['hour'] = df_local.index.hour
counts_by_hour = df_local[df_local.hour >= 0 ].groupby('hour')
counts_by_hour['_number'].count().plot(kind='bar')
```

- The first is UTC, so scewed - maybe shows european/us popularity?
- Second is in local timezones but clearly shifted later in the day
- Doesn't take into account DST.

# Theory: Each patch is reviewed more than previously.

```python
df['year'] = df.index.year
df['month'] = df.index.month
counts_by_hour = df[df['status'] == 'MERGED'].groupby(['year', 'month'])
counts_by_hour['Code-Review.total'].aggregate([np.mean, np.std]).plot()
```

- Here we just look at the merged patches to see how many reviews they got in average
- Very noisy, high std but otherwise a fairly clear trend

# Theory: Larger patches have less reviewers

```python
df[df['revisions.lines_total'] < 10000].plot(x='revisions.lines_total', y='Code-Review.total', kind='scatter')
```

# Theory: Comments are more likely to be in a negative context

```python
df.plot(x='Code-Review.all.comments', y='Code-Review.all.-1', kind='scatter')
```

```python
df.plot(x='Code-Review.all.comments', y='Code-Review.all.+1', kind='scatter')
```

- A fairly clear pattern here, the more comments you get the more -1's but not so much with +1

# Theory: OpenStack is a global project.

```python
df['owner.tz.hour'] = df['owner.tz'] / 60
counts_by_hour = df.groupby(['owner.tz.hour'])
counts_by_hour['owner._account_id'].count().plot(kind='bar')
```

# OpenStack recently but rapidly became a global project

```python
df['owner.tz.hour'] = df['owner.tz'] / 60
counts_by_hour = df.groupby(['year', 'owner.tz.hour'])
counts_by_hour['owner._account_id'].count().plot(kind='bar')
```

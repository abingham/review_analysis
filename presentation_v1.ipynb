# Bring out the data

```python
# Load the data, this is an expensive operation.
df = pd.read_csv(
    'openstack-data.csv', 
    low_memory=False, 
    parse_dates=[
        'created', 'created.local', 'merged.date', 'merged.date.local', 
        'Workflow.approved.on', 'Workflow.approved.on.local'], 
    index_col='created',
    encoding='utf-8',
)
```

```python
# Some dates are invalid in created, so force it.
df.index = pd.to_datetime(df.index, coerce=True)

# Some sanity checking. (Expect just under 120,000)
print len(df)

# Take a peek!
df[['owner.name', 'status', 'project', 'revisions.lines_total', 'revisions.commit.message']].head()
```

```python
# Average number of revisions by the review status
incubated_projects = df[df['project.group'] == 'openstack']
by_status = incubated_projects.groupby('status')
by_status['revisions.count'].aggregate([np.mean, np.std]).plot(kind='bar')
```

```python
# Theory: Larger patches require more revisions to get right
by_revisions_count = incubated_projects.groupby('revisions.count')
by_revisions_count['revisions.lines_total'].aggregate([np.mean, np.std]).plot()
```

```python
# How many reviews does a typical patch get?
reviewed = incubated_projects.groupby(['Code-Review.total'])
reviewed['owner._account_id'].count().plot()
```

```python
# How many reviews do my patches get?
df['review_duration_days'] = ((df['review_duration'] / 60) / 60) / 24
dougals = df[df['owner.name'] == 'Dougal Matthews']
dougals['review_duration_days'].plot(kind='bar')
```

```python
# Which day is the most popular for submitting reviews?
df['weekday'] = df.index.weekday
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Friday']
counts_by_day = df[df.weekday >= 0 ].groupby('weekday')
counts_by_day.index = days
counts_by_day['_number'].count().plot(kind='bar')
```

```python
# Which day is the most popular for submitting reviews? (localised)
df_local = df.set_index('created.local')
df_local['weekday'] = df_local.index.weekday
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Friday']
counts_by_day = df_local[df_local.weekday >= 0 ].groupby('weekday')
counts_by_day.index = days
counts_by_day['_number'].count().plot(kind='bar')
```

```python
# Which hour is most popular for sumitting reviews?
df['hour'] = df.index.hour
counts_by_hour = df[df.hour >= 0 ].groupby('hour')
counts_by_hour['_number'].count().plot(kind='bar')
```

```python
# Which hour is most popular for sumitting reviews?
counts_by_hour = df.groupby('owner.tz')
counts_by_hour['_number'].count().plot(kind='bar')
```

```python
# Which hour is most popular for sumitting reviews? (localised)
df_local = df.set_index('created.local')
df_local['hour'] = df_local.index.hour
counts_by_hour = df_local[df_local.hour >= 0 ].groupby('hour')
counts_by_hour['_number'].count().plot(kind='bar')
```

```python
# Which hour is most popular for approving reviews? (localised)
df_local = df.set_index('Workflow.approved.on.local')
df_local['hour'] = df_local.index.hour
counts_by_hour = df_local[df_local.hour >= 0 ].groupby('hour')
counts_by_hour['_number'].count().plot(kind='bar')
```

```python
# Are patches being reviewed by more or less people year on year?
df['year'] = df.index.year
df['month'] = df.index.month
counts_by_hour = df[df['status'] == 'MERGED'].groupby(['year', 'month'])
counts_by_hour['Code-Review.total'].aggregate([np.mean, np.std]).plot()
```

```python
# Are reviews being merged more slowly or more quickly?
df['year'] = df.index.year
df['month'] = df.index.month
df['review_duration_days'] = ((df['review_duration'] / 60) / 60) / 24
counts_by_hour = df[df['status'] == 'MERGED'].groupby(['year', 'month'])
counts_by_hour['review_duration_days'].aggregate([np.mean, np.std]).plot()
```

```python
# Are reviews being merged more slowly if you submit them as specific times?
df['hour'] = df.index.hour
counts_by_hour = df[df['hour'] > 0].groupby('hour')
counts_by_hour['review_duration_days'].aggregate([np.mean, np.std]).plot(kind='bar')
```

```python
# Are reviews being merged more slowly if you submit them as specific times? (localised)
df_local = df.set_index('created.local')
df_local['hour'] = df_local.index.hour
counts_by_hour = df_local[df_local['hour'] > 0].groupby('hour')
counts_by_hour['review_duration_days'].aggregate([np.mean, np.std]).plot(kind='bar')
```

```python
# Are reviews requiring more updates (patch sets) over time to get merged?
df['year'] = df.index.year
df['month'] = df.index.month
counts_by_hour = df[df['year'] > 0].groupby(['year', 'month'])
counts_by_hour['revisions.count'].aggregate([np.mean, np.std]).plot()
```
